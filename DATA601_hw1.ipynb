```python
#import pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#upload dataset
happiness = pd.read_csv("data/happiness.csv", nrows = 25)
#since this is a fairly large dataset I used nrows to only import the top 25 countries
```

```python
#display first 25 rows of dataset
happiness.head(25)
```

```python
#checking to see if any columns are empty
happiness.info()
#there are no missing values
```

```python
#print column names
print(happiness.columns.tolist())
```

```python
#take out spaces and save column names
col_names = ['Overall_Rank', 'Country_Or_Region', 'Score', 'Gdp_Per_Capita',
       'Social_Support', 'Healthy_Life_Expectancy',
       'Freedom_To_Make_Life_Choices', 'Generosity',
       'Perceptions_Of_Corruption']

happiness.columns = col_names
```

```python
#checking for outliers in the GDP Per Capita category for top 25 countries
happiness.Gdp_Per_Capita.plot(kind = 'box')
#The outliers appear to be Mexico and Costa Rica on the low end and Luxembourg on the high end
```

```python
#checking for outliers in the Social Support Category category for top 25 countries
happiness.Social_Support.plot(kind = 'box')
#The outlier appears to be United Arab Emirates
```

```python
#checking for outliers in the Freedom To Make Life Choices Category category for top 25 countries
happiness.Freedom_To_Make_Life_Choices.plot(kind = 'box')
```

```python
#checking for outliers in the Perceptions of Corruption category for top 25 countries
happiness.Perceptions_Of_Corruption.plot(kind = 'box')
```

```python
#checking for outliers in the Healthy Life Expectancy category for top 25 countries
happiness.Healthy_Life_Expectancy.plot(kind = 'box')
#The United States, United Arab Emirates, Czech Republic, Mexico, and Taiwan appear to be the outliers
```

```python
#checking for outliers in the Genorosity category for top 25 countries
happiness.Generosity.plot(kind = 'box')
```

```python
#Checking the mean Score
happiness.Score.mean()
```

```python
#creating bar graph that displays the happiness scores of the top 25 countries
plt.bar("Country_Or_Region", "Score", data = happiness, color = "pink")
plt.ylabel("Score")
plt.xlabel("Country Or Region")
plt.xticks(rotation = 90)
plt.title("Countries Overall Happiness Score 2019")
plt.show()
```
